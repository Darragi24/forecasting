```python
import statsmodels.tsa.stattools as ts
```

```python
average_price = data['Average Price']
result = ts.adfuller(average_price)

print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

if result[1] <= 0.05:
    print("Time series is likely stationary.")
else:
    print("Time series is likely non-stationary.")
```

```python
print(data.isnull().sum())  
```

```python
data.dropna(inplace=True)
```

```python
data['Average_Price_Diff1'] = average_price.diff().dropna()
result = ts.adfuller(data['Average_Price_Diff1'])
print("\nFirst-order Differencing:")
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
if result[1] <= 0.05:
    print("Time series is likely stationary.")
else:
    print("Time series is likely non-stationary.")
```

```python
import statsmodels.api as sm
from statsmodels.tsa.arima.model import ARIMA
from statsmodels.tsa.stattools import adfuller
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
differenced_data = data['Average_Price_Diff1']

#Check if it is stationary
result = adfuller(differenced_data)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
plt.figure(figsize=(12, 6))
plot_acf(differenced_data, lags=40)
plt.title('ACF Plot of Differenced Data')
plt.show()

plt.figure(figsize=(12, 6))
plot_pacf(differenced_data, lags=40)
plt.title('PACF Plot of Differenced Data')
plt.show()
```

```python
model_10 = ARIMA(differenced_data, order=(1, 1, 0))
model_fit_10 = model_10.fit()
print("ARIMA(1, 1, 0) Model Summary:")
print(model_fit_10.summary())
aic_10 = model_fit_10.aic
bic_10 = model_fit_10.bic
print(f"AIC: {aic_10:.2f}, BIC: {bic_10:.2f}\n")


# Fit ARIMA(1, 1, 1) model
model_11 = ARIMA(differenced_data, order=(1, 1, 1))
model_fit_11 = model_11.fit()
print("ARIMA(1, 1, 1) Model Summary:")
print(model_fit_11.summary())
aic_11 = model_fit_11.aic
bic_11 = model_fit_11.bic
print(f"AIC: {aic_11:.2f}, BIC: {bic_11:.2f}\n")


# Compare AIC and BIC
print("Model Comparison:")
print(f"ARIMA(1, 1, 0): AIC={aic_10:.2f}, BIC={bic_10:.2f}")
print(f"ARIMA(1, 1, 1): AIC={aic_11:.2f}, BIC={bic_11:.2f}")

if aic_10 < aic_11:
    print("\nARIMA(1, 1, 0) has a lower AIC, suggesting it might be a better fit.")
elif aic_11 < aic_10:
    print("\nARIMA(1, 1, 1) has a lower AIC, suggesting it might be a better fit.")
else:
    print("\nAIC values are very similar.")

if bic_10 < bic_11:
    print("\nARIMA(1, 1, 0) has a lower BIC, suggesting it might be a better fit.")
elif bic_11 < bic_10:
    print("\nARIMA(1, 1, 1) has a lower BIC, suggesting it might be a better fit.")
else:
    print("\nBIC values are very similar.")
```

```python
# Forecast the differenced data (e.g., for the next 12 periods)
forecast_steps = 365
forecast_diff = model_fit_11.predict(start=len(differenced_data), end=len(differenced_data) + forecast_steps - 1)
```

```python
# Inverse differencing to get forecasts in original scale
last_value = data['Average Price'].iloc[-1]
forecast = forecast_diff.cumsum() + last_value
```

```python
# Create a NEW DataFrame for forecasts with correct index
forecast_df = pd.DataFrame({'Forecast': forecast})
forecast_df.index = pd.date_range(start=data.index[-1] + pd.Timedelta(days=1), periods=forecast_steps, freq='D')
```

```python
#Plot Results
plt.figure(figsize=(12,6))
plt.plot(data['Average Price'], label='Actual Average Price')
plt.plot(forecast_df['Forecast'], label='Forecast Average Price') #plot the forecast dataframe
plt.legend()
plt.title('Average Price Forecast')
plt.show()

print("Forecasts:\n", forecast_df)
```

```python
scaler = MinMaxScaler(feature_range=(-1, 1))
scaled_data = scaler.fit_transform(data)

seq_length = 30

x = []
y = []
for i in range(len(scaled_data) - seq_length):
    x.append(scaled_data[i:i + seq_length])
    y.append(scaled_data[i + seq_length])
x = torch.tensor(np.array(x), dtype=torch.float32)
y = torch.tensor(np.array(y), dtype=torch.float32).reshape(-1, 1)

train_size = int(len(x) * 0.8)
x_train, x_test = x[:train_size], x[train_size:]
y_train, y_test = y[:train_size], y[train_size:]

input_size = 1
hidden_size = 64 
output_size = 1
num_epochs = 100
learning_rate = 0.001
dropout_prob = 0.2
class LSTM(nn.Module):
    def __init__(self, input_size, hidden_size, num_layers, output_size,dropout_prob):
        super().__init__()
        self.lstm = nn.LSTM(input_size, hidden_size, num_layers, batch_first=True,dropout=dropout_prob)
        self.linear = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        out, _ = self.lstm(x)
        return self.linear(out[:, -1, :])

num_layers = 3 
model = LSTM(input_size, hidden_size, num_layers, output_size,dropout_prob)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)

# Training loop (no early stopping)
for epoch in range(num_epochs):
    optimizer.zero_grad()
    outputs = model(x_train)
    loss = criterion(outputs, y_train)
    loss.backward()
    optimizer.step()
    if (epoch + 1) % 10 == 0:
        print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {loss.item():.4f}')

# Prediction and inverse transform (same as before)
model.eval()
with torch.no_grad():
    train_predict = model(x_train)
    test_predict = model(x_test)

train_predict = scaler.inverse_transform(train_predict.detach().numpy())
test_predict = scaler.inverse_transform(test_predict.detach().numpy())
```

```python
plt.figure(figsize=(12,6))
plt.plot(scaler.inverse_transform(y_train.detach().numpy()), label='Actual')
plt.plot(train_predict, label='Training Prediction')
plt.legend()
plt.show()

plt.figure(figsize=(12,6))
plt.plot(scaler.inverse_transform(y_test.detach().numpy()), label='Actual')
plt.plot(test_predict, label='Test Prediction')
plt.legend()
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score
mae = mean_absolute_error(y_test, test_predict)
rmse = np.sqrt(mean_squared_error(y_test, test_predict))
mape = mean_absolute_percentage_error(y_test, test_predict)
r2 = r2_score(y_test, test_predict)

print(f"MAE: {mae:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"MAPE: {mape:.2f}")
print(f"R-squared: {r2:.2f}")
```
